## Q1: Show that the correlation co-efficient is indeed related to the fraction of varaince explained by a linear dependence.

````python
**1** To start, make a $N(0,1)$ random variable $x$, with 200,000 entries.  Then make a linearly dependent variable $y_i= a x_i + n_i$, where a is some amplitude, and $n_i$ is a random noise varaible.  Assume that $n_i$ is distributed as $N(0,n)$, where $n$ is the noise level.

Calculate the correlation co-efficient $r_{xy}$ between $x$ and $y$ for 40 different values of $n$, logarithmically spaced between 0.01 and 100. (hint use `np.logspace`).  

Plot $r_{xy}$ versus the inverse of the noise $n$ normalized by the standard devaition of the signal $ax_i$: $\frac{a s_x}{n}$   You may want to use `ax.loglog` to visualize the orders of magnitude.  

Comment on the resulting comparison.

#### Coding hint:

You are going to need to generate 40 realizations of the random process, so do this in a `for`-loop, and save the results in an array that you initialize before the for-loop:
```python
r = np.zeros(40)
for ind,n in enumerate(noisevalues):
  # make your random variable
  ...
  # calculate r
  r[ind]=...
```
<hr>
````

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
import math as math
import scipy.stats as stats
r=np.zeros(40)
a=5.0
sx=0.0
noisevalues= np.logspace(-2,2,num=40)# will go from .01 to 100 for base 10.0
for ind,n in enumerate(noisevalues):
    xi=np.random.randn(200000)
    ni=np.random.normal(0,n,200000)#does this give N(0, n), where n is noise level and standard deviation 
    yi=a*xi+ni
    sx=np.std(xi, ddof=1)
    rxy=np.sum((xi-np.mean(xi))*(yi-np.mean(yi)))/sx/np.std(yi, ddof=1)/(200000-1)
    r[ind]=rxy
    
xAxis=a*sx/noisevalues   

fig= plt.figure(figsize=(6,10))
print fig
ax= fig.add_subplot(2,1,1)
ax.plot(xAxis, r )
ax.set_title('Correlation co-efficient vs. inverse of Noise')
ax.set_ylabel('correlation coefficients')
ax.set_xlabel('Inverse of noise normalized by signal\'s standard deviation')
ax.set_xlim([-50.0,500])
ax.set_ylim([0, 1.1])

ax= fig.add_subplot(2,1,2)
ax.loglog(xAxis,r)
ax.set_title('Log of Correlation co-efficient vs. Log of inverse of Noise')
ax.set_ylabel('Log (correlation coefficients)')
ax.set_xlabel('Log (Inverse of noise normalized by signal\'s standard deviation)')
ax.set_xlim([math.pow(10,-1.5),math.pow(10,3)])
ax.set_ylim([math.pow(10,-1.5), 1.5])
````

From the graphs, we can see that as noise is very high(ie, inverse noise is very low), the correlation is small. But as noise decreases, thus increasing its inverse, the correlation approaches 1, thus becomes strongly correlated. This is much easier to see on the loglog graph.

## Q2: Monte Carlo on slopes and test versus student-t distribution

**1** We stated without proof that the slope $b$ of a linear fit of $y$ to $x$ is distributed as a student-t distribution with $N-2$ degrees of freedom, with a standard devaition given by the standard deviation of $y$ around the fit model $\tilde{y}$.   Lets test that emperically using a Monte Carlo technique.  

First create an independet variable $x$ from 10 points from an $N(0,1)$ distribution.

Next, using $n=1.0$, and $a=1.0$ create the dependent variables $y_i=a x_i + n_i$ where $n_i$ are drawn from the $N(0,n)$ distribution.  Calculate the slope $b$.  

Do this 10000 times, and make a histogram of the resulting fits $b$.  Also histogram $s_{y|x}$ as defined in the notes.

Comment on the histigrams

<hr>

````python
N=10000
bvals=np.zeros(N)
syxvals=np.zeros(N)
for ind,n in enumerate(bvals):
    xi=np.random.randn(10)
    a=1.0
    n=1.0
    ni=np.random.normal(0,n,10)
    yi=a*xi+ni
    #slope= b
    b= np.sum(((yi-np.mean(yi))*xi))/np.sum((xi-np.mean(xi))*xi)
    bvals[ind]=b
    anew= np.mean(yi)-b*np.mean(xi)
    yhat=b*xi+anew
    syx= np.sqrt(np.sum((yi-yhat)**2)/(N-2))
    syxvals[ind]=syx
fig=plt.figure(figsize=(12,7))
ax=fig.add_subplot(1,2,1)
ax.hist(bvals)
ax.set_xlabel('b values, the slope')
ax.set_ylabel('Occurences')
ax.set_title('Distribution of the slopes')
print 'The Skew of the slope values:%1.3f'%stats.skew(bvals)

ax=fig.add_subplot(1,2,2)
ax.hist(syxvals)
ax.set_xlabel('$s_{y|x}$')
ax.set_ylabel('Occurances')
ax.set_title('Standard Deviation of the slopes')
print 'The Skew of the $s_{y|x}$ values:%1.3f'%stats.skew(syxvals)
    
````

The histogram of the slopes certainly does have the general shape of a normal distrubution, which the student-t distrubution looks like for large N. The histogram of the standard deviation shows a distribution that also appears to be normal, though the "slant" of it is opposite to the fist plot- it seems to have more of a rightward leaning, and the plot of the slope has more of a leftward leaning. This can also be seen by the above calculated values of the Skew. 


**2** Assuming $s_{y|x}=1.$ (which is the same as our value of $n$), plot the student-t probability density function on the histogram of b and show that they compare favorably (if they don't you might need to check your work!)

<hr>

````python
x= np.linspace(-7., 7., 1000)
ndeg= N-2
y= stats.t.pdf(x,ndeg)
fig=plt.figure(figsize=(5,5))
ax=fig.add_subplot(1,1,1)
ax.hist((bvals-np.mean(bvals))/np.std(bvals),50, normed=True, label=('hist of slopes'))# to normalize distrubution and centre it at zero
ax.plot(x,y,'r--', lw=2, label=('Student-t prob, dof=%2d' %ndeg))
ax.set_xlabel('b values, the slope')
ax.set_ylabel('Occurences')
ax.set_title('Normed Slope Distrubution with Student-t Probability')
ax.set_xlim(-7.0, 7.)
ax.legend(loc=0)

````

````python
fig= plt.figure(figsize=(7,7))
ax=fig.add_subplot(1,1,1)
ax.plot(dc,jb,'.', markersize=2)
ax.set_xlabel('Deep Cove')
ax.set_ylabel('James Bay')
ax.set_title('Temperature Data from James Bay and Deep Cove ')
````

**2** Compute the linear Correlation Co-efficient (without a canned routine)

````python
N=len(jb)
sx=np.std(dc, ddof=1)
sy=np.std(jb,ddof=1)
rxy=np.sum((dc-np.mean(dc))*(jb-np.mean(jb)))/sx/sy/(N-1)
print 'Linear Correlation Coefficient: %1.4f' %rxy
````

**3** Fit a straight line to the James Bay data as a fucntion of the Deep Cove data, and add to the scatter plot made above.  Comment on the fit.

<hr>

````python
b=np.sum(dc*jb)-N*np.mean(dc)*np.mean(jb)
b= b/(np.sum(dc**2)-N*np.mean(dc)**2)
a= np.mean(jb)-b*np.mean(dc)
yhat= b*dc+a

fig= plt.figure(figsize=(7,7))
ax=fig.add_subplot(1,1,1)
ax.plot(dc,jb,'.')
ax.plot(dc, yhat, '-', label='Line of best fit: y=%1.4f+%1.4fx'%(a,b))
ax.set_xlabel('Deep Cove')
ax.set_ylabel('James Bay')
ax.set_title('Temperature Data from James Bay and Deep Cove ')
ax.legend(loc=0)
````

The fit is fairly well, though at either end of the points, the line does seem to be a bit off. And it looks like a murmuration of birds!

**4** Add the confidence intervals to your fit, and comment.

````python
b=np.sum(dc*jb)-N*np.mean(dc)*np.mean(jb)
b= b/(np.sum(dc**2)-N*np.mean(dc)**2)
a= np.mean(jb)-b*np.mean(dc)
yhat= b*dc+a
syx= np.sqrt(np.sum((jb-yhat)**2)/(N-2))
bdn,bup=stats.t.interval(0.95,df=N-2,scale=syx)
bb=np.array([bdn,bup])
bb=bb/np.sqrt(np.sum((dc-np.mean(dc))**2))
print 'Slope b is between %f and %f with 95%% probability'%(b+bb[0],b+bb[1])
xx=np.sort(dc)
yhathigh=(b+bb[1])*(xx-np.mean(dc))+np.mean(jb)
yhatlow=(b+bb[0])*(xx-np.mean(dc))+np.mean(jb)

fig= plt.figure(figsize=(10,10))
ax=fig.add_subplot(2,1,1)
ax.plot(dc,jb,'.', markersize=2)
ax.plot(dc, yhat, '-', label='Line of best fit: y=%1.4f+%1.4fx'%(a,b))
ax.set_xlabel('Deep Cove')
ax.set_ylabel('James Bay')
ax.set_title('Temperature Data from James Bay and Deep Cove, with scale=$s_{y|x}$ ')
ax.legend(loc=0)
ax.fill_between(xx,yhathigh,yhatlow,alpha=0.4)
ax.set_xlim([-10.,40.])
ax.set_ylim([-10.,35.])
#print second plot
bdn,bup=stats.t.interval(0.95,df=N-2,scale=50)
bb=np.array([bdn,bup])
bb=bb/np.sqrt(np.sum((dc-np.mean(dc))**2))
xx=np.sort(dc)
yhathigh=(b+bb[1])*(xx-np.mean(dc))+np.mean(jb)
yhatlow=(b+bb[0])*(xx-np.mean(dc))+np.mean(jb)

ax=fig.add_subplot(2,1,2)
ax.plot(dc,jb,'.', markersize=2)
ax.plot(dc, yhat, '-', label='Line of best fit: y=%1.4f+%1.4fx'%(a,b))
ax.set_xlabel('Deep Cove')
ax.set_ylabel('James Bay')
ax.set_title('Temperature Data from James Bay and Deep Cove, with Scale=50 ')
ax.legend(loc=0)
ax.fill_between(xx,yhathigh,yhatlow,alpha=0.4)
ax.set_xlim([-10.,40.])
ax.set_ylim([-10.,35.])
````

**NOTE** There is a way to treat both axes independently called "Neutral Regression".  It is relatively simple, and involves normalizing the total distance of each point from the trend line. <a href="http://journals.ametsoc.org/doi/pdf/10.1175/1520-0426(1999)016%3C0876%3AAPFANR%3E2.0.CO%3B2">(Marsden 1999)</a>. This technique is strongly preferred if there is no reason to think that one data set is dependent on the other.  A caveat is that the data sets shoudl be normalized before fitting so that the "distance" has meaning.
